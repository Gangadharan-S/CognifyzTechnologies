```python
#Importing All Necessary Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.info()
```

Create a histogram or bar chart to
visualize the distribution of price ranges
among the restaurants.

```python
plt.subplots(figsize=(10,6))
plt.hist(df['Price range'],bins=df['Price range'].nunique(),edgecolor='k')
plt.title('Price Range Distribution')
plt.xlabel('Price Range')
plt.ylabel('Frequency')
plt.show()
```

Calculate the percentage of restaurants
in each price range category.

```python
#creates a dataframe which holds the counts of restaurents by each  price range:

Price_range_counts  =df['Price range'].value_counts().reset_index()
Price_range_counts
```

```python
# Create a DataFrame to hold the counts of restaurants in each price range
price_range_counts = df['Price range'].value_counts().reset_index()

# Calculate the total number of restaurants
total_restaurants = len(df)

# Calculate the percentage of restaurants in each price range category
# Use the 'normalize' argument in value_counts to get percentages directly
percentage_per_price_range = df['Price range'].value_counts(normalize=True) * 100

# Print result
print("Percentage of restaurants in each price range category:")
for price_range, percentage in percentage_per_price_range.items():
    print(f"{price_range}: {percentage:.2f}%")
```

```python
# Create the figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the bar chart
percentage_per_price_range.plot(kind='bar', ax=ax, color='fuchsia', edgecolor='k')

# Annotate bars with their values
for bar in ax.containers[0]:
    height = bar.get_height()
    ax.annotate(
        f'{height:.2f}%',
        xy=(bar.get_x() + bar.get_width() / 2, height),
        xytext=(0, 3),
        textcoords='offset points',
        ha='center',
        va='bottom'
    )

# Set the title and labels
ax.set_title('Percentage of Restaurants in Each Price Range')
ax.set_xlabel('Price Range')
ax.set_ylabel('Frequency')
plt.show()# Display the plot
```

**Task-4 Restaurant Chains**

Identify if there are any restaurant chains
present in the dataset.

```python
# Group by 'Restaurant Name' and count the number of outlets
restaurant_chain = df.groupby("Restaurant Name").size().reset_index(name="OutletCount")

# Filter for restaurants with more than one outlet
filtered_chains = restaurant_chain[restaurant_chain["OutletCount"] > 1]

# Sort the results by 'OutletCount' in descending order
restaurant_chains = filtered_chains.sort_values(by="OutletCount", ascending=False)
```

```python
print(restaurant_chains [["Restaurant Name", "OutletCount"]].head(10))
```

```python
Top_5_chains = restaurant_chains[:5]
# Create a bar plot for the top 5 restaurant chains
plt.bar(Top_5_chains["Restaurant Name"], Top_5_chains["OutletCount"], color='cyan', edgecolor='k')
# Add labels and title
plt.xlabel("Restaurant Chain")
plt.ylabel("Number of Outlets")
plt.title("Top 5 Restaurant Chains by Number of Outlets")
plt.show()
```

Analyze the ratings and popularity of
different restaurant chains.

```python
# Calculate the mean aggregate rating for each restaurant
ratings = (df.groupby("Restaurant Name")["Aggregate rating"].mean().reset_index(name="Average Rating").sort_values(by="Average Rating", ascending=False))
print(ratings)
```

```python
votes = (df.groupby("Restaurant Name")["Votes"].sum().reset_index(name="Total Votes").sort_values(by="Total Votes", ascending=False))
print(votes)
```

```python
top_5_ratings = ratings.head(5)
colors = ['turquoise', 'teal', 'aqua', 'darkgreen', 'darkorange']
plt.figure(figsize=(10, 6))
plt.bar(x=top_5_ratings["Restaurant Name"], height=top_5_ratings["Average Rating"],color=colors,edgecolor='k')
plt.xlabel("Restaurant Chain", fontsize=12)
plt.ylabel("Average Rating", fontsize=12)
plt.title("Top 5 Restaurant Chains by Average Rating", fontsize=14)
plt.tight_layout()
plt.show()
```

**Task-1 Restaurant Reviews**

Analyze the text reviews to identify the most
common positive and negative keywords.

```python
print(df['Rating text'])
```

```python
positive_data = df.loc[df['Aggregate rating'] >= 2.5]
print(positive_data)
```

```python
positive_keywords = positive_data['Rating text'].unique()
print(positive_keywords)
```

```python
from collections import Counter

keyword_counts = Counter()
for text in positive_data['Rating text']:
    for keyword in positive_keywords:
        if keyword in text:
            keyword_counts[keyword] += 1

# Extract keywords and their frequencies
keywords = list(keyword_counts.keys())
frequencies = list(keyword_counts.values())

# Plot the histogram (bar chart in this case)
plt.bar(keywords, frequencies, color='indigo', edgecolor='k')
plt.xlabel("Positive Keywords")
plt.ylabel("Frequency")
plt.title("Frequency of Positive Keywords in Ratings")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
negative_data = df.loc[(df['Aggregate rating'] < 2.5) & (df['Aggregate rating'] > 0)]
#print(negative_data)
negative_keywords = negative_data['Rating text'].unique()
print(negative_keywords)
```

```python
plt.hist(negative_data['Rating text'], label = negative_keywords)
plt.show()
```

**Task-3 Price Range vs Online Delivery and Table Booking.**

Analyze if there is a relationship between the
price range and the availability of online
delivery and table booking.

```python
x1 = df['Has Online delivery']
y = df['Average Cost for two']
x2 = df['Has Table booking']
```

```python
plt.figure(figsize=(12, 5))
# Subplot 1: Online delivery vs Average Cost for two
plt.subplot(1, 2, 1)
sns.boxplot(x=x1, y=y, palette='pastel')
plt.xlabel('Online Delivery', fontsize=12)
plt.ylabel('Average Cost for Two', fontsize=12)
plt.title('Impact of Online Delivery on Cost', fontsize=14)
# Subplot 2: Table booking vs Average Cost for two
plt.subplot(1, 2, 2)
sns.boxplot(x=x2, y=y, palette='spring')
plt.xlabel('Table Booking', fontsize=12)
plt.ylabel('Average Cost for Two', fontsize=12)
plt.title('Impact of Table Booking on Cost', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
print(" Both the Online delivery and Table booking are negetively impacted by the cost")
```

Determine if higher-priced restaurants are
more likely to offer these services.

```python
print('High priced Restarants which have Average cost More than 10000 not provided online delivary as will as Table booking')
```

```python
plt.figure(figsize = (9,4))
plt.subplot(1,2,1)
sns.boxplot(x=x1,y = 'Average Cost for two',data = df.loc[df['Average Cost for two']<10000])
plt.title('Online delivery by cost')
plt.subplot(1,2,2)
sns.boxplot(x = x2, y = 'Average Cost for two',data = df.loc[df["Average Cost for two"]<10000])
plt.title('Table booking by cost')
plt.tight_layout()
plt.show()
```
